Carregue a imagem "pollen.tif", mostre o seu histograma, aplique uma equalização e apresente a imagem equalizada e o novo histograma. Salve o resultado da imagem com contraste aprimorado.

Certainly! Here's an example code that loads the "pollen.tif" image, displays its histogram, applies histogram equalization, shows the equalized image and its histogram, and then saves the result with enhanced contrast:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('pollen.tif', cv2.IMREAD_GRAYSCALE)

# Display the histogram of the original image
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.hist(image.ravel(), bins=256, range=[0, 256], density=True, color='blue', alpha=0.7)
plt.title('Original Histogram')

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the equalized image and its histogram
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

plt.subplot(2, 2, 4)
plt.hist(equalized_image.ravel(), bins=256, range=[0, 256], density=True, color='blue', alpha=0.7)
plt.title('Equalized Histogram')

plt.tight_layout()
plt.show()

# Save the equalized image with enhanced contrast
cv2.imwrite('enhanced_pollen.tif', equalized_image)
print("Enhanced image saved.")
```

In this code:

- We load the "pollen.tif" image using `cv2.imread()` with the `cv2.IMREAD_GRAYSCALE` flag to read it in grayscale mode.
- The histogram of the original image is displayed using `plt.hist()`.
- Histogram equalization is applied using `cv2.equalizeHist()`.
- The equalized image and its histogram are displayed.
- The equalized image is saved using `cv2.imwrite()` as "enhanced_pollen.tif".

This code provides a step-by-step demonstration of loading, displaying histograms, applying histogram equalization, and saving the result with enhanced contrast. Remember that histogram equalization can sometimes lead to overly enhanced images, so it's a good practice to adjust the degree of enhancement based on the specific image and desired outcome.

(a) Gere uma imagem 640x480 contendo níveis de cinza aleatórios com distribuição de probabilidade normal com média 128 e desvio padrão 20. Utilize semente aleatória padrão igual a 0 (zero).

Certainly! Here's an example code that generates a grayscale image of size 640x480 with randomly distributed pixel values following a normal distribution with a mean of 128 and a standard deviation of 20. The random seed is set to 0 (zero) for reproducibility:

```python
import cv2
import numpy as np

# Set random seed for reproducibility
np.random.seed(0)

# Image dimensions
height = 480
width = 640

# Generate random pixel values with normal distribution
mean = 128
stddev = 20
random_pixels = np.random.normal(mean, stddev, (height, width))

# Clip pixel values to [0, 255] range
random_pixels = np.clip(random_pixels, 0, 255)

# Convert to uint8 data type
random_pixels = random_pixels.astype(np.uint8)

# Create and display the random grayscale image
cv2.imshow('Random Grayscale Image', random_pixels)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

In this code:

- We set the random seed to 0 using `np.random.seed(0)` to ensure reproducibility.
- The image dimensions are specified as `height` and `width`.
- Random pixel values are generated using `np.random.normal()` with the specified mean and standard deviation.
- Pixel values are clipped to the [0, 255] range using `np.clip()` to ensure they are valid intensities.
- The pixel values are converted to the `uint8` data type using `.astype(np.uint8)`.
- The `cv2.imshow()` function displays the generated grayscale image.

Leia a imagem "septagon.tif", crie e adicione ruído do tipo sal e pimenta a esta imagem. Salve o resultado.

Certainly! Here's an example code that reads the "septagon.tif" image, adds salt-and-pepper noise to it, and then saves the resulting image:

```python
import cv2
import numpy as np

# Read the original image
original_image = cv2.imread('septagon.tif', cv2.IMREAD_GRAYSCALE)

# Image dimensions
height, width = original_image.shape

# Probability of adding salt-and-pepper noise
noise_prob = 0.05

# Create salt-and-pepper noise mask
noise_mask = np.random.rand(height, width)
salt_pixels = noise_mask < noise_prob / 2
pepper_pixels = noise_mask > 1 - noise_prob / 2

# Apply salt-and-pepper noise
noisy_image = np.copy(original_image)
noisy_image[salt_pixels] = 255  # Set salt pixels to white
noisy_image[pepper_pixels] = 0   # Set pepper pixels to black

# Save the noisy image
cv2.imwrite('noisy_septagon.tif', noisy_image)
print("Noisy image saved.")
```

In this code:

- We read the "septagon.tif" image using `cv2.imread()` with the `cv2.IMREAD_GRAYSCALE` flag.
- The dimensions of the image are obtained using `.shape`.
- The `noise_prob` parameter determines the probability of adding salt-and-pepper noise to each pixel.
- A random noise mask is generated using `np.random.rand()` with values between 0 and 1.
- Salt pixels are identified as pixels where the noise mask is below `noise_prob / 2`, and pepper pixels are identified as pixels where the noise mask is above `1 - noise_prob / 2`.
- The noisy image is created by copying the original image and replacing salt pixels with white (255) and pepper pixels with black (0).
- The noisy image is saved using `cv2.imwrite()` as "noisy_septagon.tif".

This code generates an image with salt-and-pepper noise and saves it. Remember to have the OpenCV library installed (`pip install opencv-python`) to run this code successfully.

Certainly! Here's the code to perform spatial domain filtering on the "ckt.tif" image to obtain the gradients gx and gy in the x and y directions, respectively, and save the results:

```python
import cv2
import numpy as np

# Load the image "ckt.tif"
ckt_image = cv2.imread('ckt.tif', cv2.IMREAD_GRAYSCALE)

# Apply Sobel filters to obtain gradients gx and gy
sobel_dx = cv2.Sobel(ckt_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_dy = cv2.Sobel(ckt_image, cv2.CV_64F, 0, 1, ksize=3)

# Calculate the absolute gradients gx and gy
gx = np.abs(sobel_dx)
gy = np.abs(sobel_dy)

# Save the results
cv2.imwrite('gx_image.tif', gx)
cv2.imwrite('gy_image.tif', gy)

print("Results saved.")
```

In this code:

- We load the "ckt.tif" image using `cv2.imread()` in grayscale mode.
- We apply the Sobel filters using `cv2.Sobel()` to obtain gradients `sobel_dx` and `sobel_dy` in the x and y directions, respectively.
- We calculate the absolute gradients `gx` and `gy` by taking the absolute values of `sobel_dx` and `sobel_dy`.
- The resulting `gx` and `gy` images are saved as 'gx_image.tif' and 'gy_image.tif'.

This approach helps in identifying the intensity changes in the x and y directions of the image, which are useful for various image processing tasks, including edge detection and feature extraction.

Certainly! Here's the code to apply frequency domain filtering to attenuate periodic noise in the "lena_noise.png" image and save the result:

```python
import cv2
import numpy as np
from numpy.fft import fft2, fftshift, ifft2, ifftshift

# Load the noisy Lena image
lena_noise_image = cv2.imread('lena_noise.png', cv2.IMREAD_GRAYSCALE)

# Compute the 2D Discrete Fourier Transform of the noisy image
lena_noise_fft = fft2(lena_noise_image)

# Shift the zero frequency component to the center
lena_noise_fft_shifted = fftshift(lena_noise_fft)

# Design a high-pass filter to attenuate periodic noise
rows, cols = lena_noise_image.shape
center_row, center_col = rows // 2, cols // 2
cutoff_frequency = 50  # Adjust this value based on the noise frequency
high_pass_filter = np.ones_like(lena_noise_fft_shifted)
high_pass_filter[center_row - cutoff_frequency:center_row + cutoff_frequency,
                 center_col - cutoff_frequency:center_col + cutoff_frequency] = 0

# Apply the high-pass filter in the frequency domain
filtered_fft_shifted = lena_noise_fft_shifted * high_pass_filter

# Shift the frequency domain back to the original position
filtered_fft = ifftshift(filtered_fft_shifted)

# Compute the inverse 2D Discrete Fourier Transform
filtered_image = np.abs(ifft2(filtered_fft))

# Normalize the filtered image to 8-bit range
filtered_image = cv2.normalize(filtered_image, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

# Save the filtered image
cv2.imwrite('filtered_lena.png', filtered_image)

print("Filtered image saved.")
```

In this code:

- We load the "lena_noise.png" image using `cv2.imread()` in grayscale mode.
- We compute the 2D Discrete Fourier Transform of the noisy image using `fft2()`.
- The zero frequency component is shifted to the center using `fftshift()`.
- We design a high-pass filter to attenuate periodic noise by setting the center frequencies to zero while keeping the noise frequency components.
- The high-pass filter is applied to the frequency domain image.
- We shift the frequency domain back to the original position using `ifftshift()`.
- We compute the inverse 2D Discrete Fourier Transform using `ifft2()` to obtain the filtered image.
- The filtered image is normalized to the 8-bit range using `cv2.normalize()`.
- The filtered image is saved as 'filtered_lena.png'.

Frequency domain filtering is effective for attenuating specific frequencies or patterns, making it useful for removing periodic noise. Remember to have the OpenCV library installed (`pip install opencv-python`) to run this code successfully.
