## Define image paths and filenames

```python
PATH = "../img/"
LENA = os.path.join(PATH, "lena.png")
```

```python
x = np.uint8(250)
y = np.uint8(10)
z = x + y
print(x)
print(y)
print(z)
```

Now let's see how it works using OpenCV

```python
# The foolowing snippet results in error in OpenCV:
# error: OpenCV(4.5.2) :-1: error: (-5:Bad argument) in function 'add'
# > Overload resolution failed:
# >  - src1 is not a numpy array, neither a scalar
# >  - Expected Ptr<cv::UMat> for argument 'src1'
x = np.uint8(250)
y = np.uint8(10)
z = cv2.add(x, y)
print(x)
print(y)
print(z)
```

```python
# To fix it, use the following
x = np.uint8([250])
y = np.uint8([10])
z = cv2.add(x, y)
print(x)
print(y)
print(z)
```

## Add a scalar to an image

```python
img = cv2.imread(LENA, cv2.IMREAD_COLOR)
val = 50
img2 = img + val
img3 = cv2.add(img, val)
```

```python
plt.figure(figsize=(18,18))
plt.subplot(221); plt.title("IMG"); plt.imshow(bgr2rgb(img))
plt.subplot(222); plt.title("IMG 2"); plt.imshow(bgr2rgb(img2))
plt.subplot(223); plt.title("IMG"); plt.imshow(bgr2rgb(img))
plt.subplot(224); plt.title("IMG 3"); plt.imshow(bgr2rgb(img3))
plt.show()
```

```python
cv2.imshow("img", img)
cv2.imshow("img2", img2)
cv2.imshow("img3", img3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
