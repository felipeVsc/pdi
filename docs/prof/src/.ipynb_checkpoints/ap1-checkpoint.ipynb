$
\vec{\nabla} f(x,y) = \dfrac{\partial f(x,y)}{\partial x} + \dfrac{\partial f(x,y)}{\partial y}
$

nas direções $x$ e $y$, e a magnitude

$
\left | \vec{\nabla} f(x,y) \right | \approx \left |\dfrac{\partial f(x,y)}{\partial x}\right | + \left | \dfrac{\partial f(x,y)}{\partial y} \right |

# Solução

````python
# Solution
def me(a, b):
    return np.max(np.abs(a-b))

def mae(a, b):
    return (1 / (a.size * b.size)) * np.sum(np.abs(a - b))

def mse(a, b):
    return (1 / (a.size * b.size)) * np.sum(np.square(a - b))

def jaccard(a, b, delta = 0.2):
    return (1 / (a.size * b.size)) * np.count_nonzero(np.abs(a-b) > delta)

def log_transform(r):
    c = 255/np.log(1 + 255)
    return c * np.log(1 + r)

def inverse_log(s):
    c = 255/np.log(1 + 255)
    return np.exp(s)**(1/c) - 1

np.random.seed(1)
a = np.random.random((5,3))
b = np.random.random((5,3))


print(a)
print(b)
print(f'me(a,b)      = {me(a,b):.4e}')
print(f'mae(a,b)     = {mae(a,b):.4e}')
print(f'mse(a,b)     = {mse(a,b):.4e}')
print(f'jaccard(a,b) = {jaccard(a,b):.4e}')


````
