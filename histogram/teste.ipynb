```python
from skimage import exposure
import matplotlib.pyplot as plt
import cv2
import numpy as np

source_image = cv2.imread("source.jpg")
reference_image = cv2.imread("reference.jpg")

matched = exposure.match_histograms(source_image, reference_image, channel_axis=-1)



# plotagem
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

ax1.imshow(source_image)
ax1.set_title('Source')
ax2.imshow(reference_image)
ax2.set_title('Reference')
ax3.imshow(matched)
ax3.set_title('Matched')

plt.tight_layout()
plt.show()
```
