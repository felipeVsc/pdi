```python
import cv2
import numpy as np

# Set the dimensions of the image
width = 500
height = 500

# Set the parameters of the paraboloid
a = 100  # Scaling factor for the x-axis
b = 50   # Scaling factor for the y-axis

# Generate the x and y coordinates using meshgrid
x = np.linspace(-1, 1, width)
y = np.linspace(-1, 1, height)
X, Y = np.meshgrid(x, y)

# Calculate the pixel values for the paraboloid
Z = a * X**2 + b * Y**2

# Normalize the pixel values to the range [0, 255]
Z_normalized = ((Z - np.min(Z)) / (np.max(Z) - np.min(Z))) * 255

# Convert the pixel values to integers
image = Z_normalized.astype(np.uint8)

# Display the image
plt.imshow(image)
plt.show()

```

```python
import cv2
import numpy as np
import math

# Set the dimensions of the image
width = 500
height = 500

# Set the parameters for the sine wave
amplitude = 100
frequency = 10
rotation_angle_deg = 45

# Generate the x and y coordinates using meshgrid
x = np.linspace(-1, 1, width)
y = np.linspace(-1, 1, height)
X, Y = np.meshgrid(x, y)

# Calculate the rotated coordinates
rotation_angle_rad = math.radians(rotation_angle_deg)
rotated_X = X * math.cos(rotation_angle_rad) - Y * math.sin(rotation_angle_rad)
rotated_Y = X * math.sin(rotation_angle_rad) + Y * math.cos(rotation_angle_rad)

# Calculate the pixel values for the rotated sine wave
Z = amplitude * np.sin(2 * math.pi * frequency * rotated_X)

# Normalize the pixel values to the range [0, 255]
Z_normalized = ((Z - np.min(Z)) / (np.max(Z) - np.min(Z))) * 255

# Convert the pixel values to integers
image = Z_normalized.astype(np.uint8)

# Display the image
plt.imshow(image)
plt.show()
```
