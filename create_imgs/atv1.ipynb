Attention: Use numpy.linspace() and numpy.meshgrid() to generate the images.

```python
# * (1) Create the image of a paraboloid with one axis scaled (like an oval paraboloid).
import cv2
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

paraboloide = 100 * X**2 + 50 * Y**2
normalizado = cv2.normalize(paraboloide, None, 0, 255, cv2.NORM_MINMAX)
normalizado2 = np.asarray(normalizado, dtype=np.uint8)

plt.imshow(normalizado2)
plt.show()

```

```python
# * (4) Create a function that generates the image of a Gaussian optionally rotated by an angle \theta and with mx, my, sx, sy as input arguments.
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gaussian(mx, my, sx, sy, theta=0):
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    xx, yy = np.meshgrid(x, y)
    
    if theta != 0:
        x_rotated = xx * np.cos(theta) - yy * np.sin(theta) # testar
        y_rotated = xx * np.sin(theta) + yy * np.cos(theta)
        xx, yy = x_rotated, y_rotated
    
    gauss = np.exp(-(xx - mx)**2 / (2 * sx**2) - (yy - my)**2 / (2 * sy**2))
    
    normalized = cv2.normalize(gauss, None, 0, 255, cv2.NORM_MINMAX)
    normalized2 = np.asarray(normalized, dtype=np.uint8)

    plt.imshow(normalized2)
    plt.show()
    
gaussian(mx=0, my=0, sx=2, sy=3, theta=1)
```

```python

# (3) Create the image of a gaussian.
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
xx, yy = np.meshgrid(x, y)
    
gauss = np.exp(-(xx)**2 / (2) - (yy)**2 / (2))
    
plt.imshow(gauss)
plt.show()

```
